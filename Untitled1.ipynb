```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parameters
pi_param = 0.5309   # Π
beta = 0.5534       # β
theta = 0.5965      # θ
alpha = 0.1753      # α
omega = 3.0294      # ω
mu = 0.1753          # μ
k = 0.1752         # k
l = 0.5309        # l

# Initial conditions
S0 = 103
P0 = 57
R0 = 34

# Time vector: days 0..30 step 3
t = np.arange(0, 31, 3)

# rates
lam_S = alpha + mu
lam_P = alpha + theta + mu
lam_R = mu + omega

# HPM first-order coefficients
A_S = l * pi_param + theta * P0 + omega * R0 - beta * P0 * S0
A_P = beta * P0 * S0
R_eq = (k * pi_param + alpha * (S0 + P0)) / lam_R

# closed-form HPM approximations
S_t = S0 + (A_S / lam_S) * (1 - np.exp(-lam_S * t))
P_t = P0 + (A_P / lam_P) * (1 - np.exp(-lam_P * t))
R_t = R_eq + (R0 - R_eq) * np.exp(-lam_R * t)

# prepare table
df = pd.DataFrame({"t (days)": t, "S(t)": S_t, "P(t)": P_t, "R(t)": R_t})
print(df)
# plots (one chart per variable)
plt.figure(); plt.plot(t, S_t, marker='o'); plt.title("S(t)"); plt.xlabel("days"); plt.grid(True); plt.show()
plt.figure(); plt.plot(t, P_t, marker='o'); plt.title("P(t)"); plt.xlabel("days"); plt.grid(True); plt.show()
plt.figure(); plt.plot(t, R_t, marker='o'); plt.title("R(t)"); plt.xlabel("days"); plt.grid(True); plt.show()
```

```python
# Numerical simulation table for S, P, R over 30 days (3-day interval)
import numpy as np
import pandas as pd
from scipy.integrate import solve_ivp

# Parameter values (converted to rates by dividing by 100)
Pi = 53.09 / 100
beta = 55.34 / 100
theta = 59.65 / 100
alpha = 17.53 / 100
omega = 40.15 / 100
mu = 3.61 / 100
k = 0.0064

# Initial conditions
S0, P0, R0 = 110, 64, 41

# Model equations
def model(t, y):
    S, P, R = y
    dSdt = (1-k)*Pi - beta*P*S - alpha*S - mu*S + omega*R + theta*P
    dPdt = beta*P*S - (alpha + theta + mu)*P
    dRdt = k*Pi + alpha*(S+P) - (mu + omega)*R
    return [dSdt, dPdt, dRdt]

# Time points (0 to 30 days, interval 3)
t_eval = np.arange(0, 31, 3)

# Solve ODE
solution = solve_ivp(model, [0, 30], [S0, P0, R0], t_eval=t_eval)

# Extract results
S, P, R = solution.y

# Create table
table = pd.DataFrame({
    "Time (days)": t_eval,
    "Susceptible S(t)": np.round(S, 4),
    "Purchasing P(t)": np.round(P, 4),
    "Removed R(t)": np.round(R, 4)
})

table
```
