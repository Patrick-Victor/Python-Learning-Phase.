```python
# btts_29nov2025.py
import pandas as pd
import numpy as np
from math import exp

# ----- CONFIG -----
RECENT_MATCHES_PER_TEAM = 12
HOME_ADVANTAGE = 1.10
BTTS_THRESHOLD = 0.75

# -------------------
def compute_team_stats(results, recent_n=12):
    teams = {}
    all_teams = pd.concat([results['home_team'], results['away_team']]).unique()
    for team in all_teams:
        team_matches = results[(results['home_team']==team) | (results['away_team']==team)].sort_values('date', ascending=False).head(recent_n)
        if team_matches.empty:
            teams[team] = {'scored': np.nan, 'conceded': np.nan, 'matches': 0}
            continue
        scored = []
        conceded = []
        for _, r in team_matches.iterrows():
            if r['home_team'] == team:
                scored.append(r['home_goals']); conceded.append(r['away_goals'])
            else:
                scored.append(r['away_goals']); conceded.append(r['home_goals'])
        teams[team] = {'scored': np.mean(scored), 'conceded': np.mean(conceded), 'matches': len(scored)}
    return teams

def league_avg_goals(results):
    total_goals = results['home_goals'].sum() + results['away_goals'].sum()
    matches = len(results)
    return total_goals / matches if matches>0 else 2.7

def compute_strengths(results, recent_n=12):
    strengths = {}
    for league, group in results.groupby('league'):
        g_avg = league_avg_goals(group)
        teams = compute_team_stats(group, recent_n)
        g_avg_per_team = g_avg / 2.0
        league_teams = {}
        for t, s in teams.items():
            if s['matches']==0 or np.isnan(s['scored']):
                attack = 1.0; defense = 1.0
            else:
                attack = (s['scored'] / g_avg_per_team) if g_avg_per_team>0 else 1.0
                defense = (s['conceded'] / g_avg_per_team) if g_avg_per_team>0 else 1.0
            league_teams[t] = {'attack': attack, 'defense': defense, 'scored': s['scored'], 'conceded': s['conceded']}
        strengths[league] = {'g_avg': g_avg, 'teams': league_teams}
    return strengths

def expected_goals(league_info, home_team, away_team, home_adv=1.1):
    g_avg = league_info['g_avg']
    teams = league_info['teams']
    a_h = teams.get(home_team, {}).get('attack', 1.0)
    d_h = teams.get(home_team, {}).get('defense', 1.0)
    a_a = teams.get(away_team, {}).get('attack', 1.0)
    d_a = teams.get(away_team, {}).get('defense', 1.0)
    lam_h = (g_avg/2.0) * a_h * d_a * home_adv
    lam_a = (g_avg/2.0) * a_a * d_h
    return lam_h, lam_a

def btts_prob(lh, la):
    p0h = exp(-lh); p0a = exp(-la); p00 = exp(-(lh+la))
    return 1.0 - p0h - p0a + p00

def run(results_path='results.csv', fixtures_path='fixtures.csv'):
    results = pd.read_csv(results_path, parse_dates=['date'])
    fixtures = pd.read_csv(fixtures_path, parse_dates=['date'])
    strengths = compute_strengths(results, RECENT_MATCHES_PER_TEAM)
    rows=[]
    for _, f in fixtures.iterrows():
        league = f.get('league', list(strengths.keys())[0] if strengths else 'LEAGUE')
        league_info = strengths.get(league, {'g_avg':2.7,'teams':{}})
        lam_h, lam_a = expected_goals(league_info, f['home_team'], f['away_team'], HOME_ADVANTAGE)
        p = btts_prob(lam_h, lam_a)
        rows.append({
            'date': f['date'].date(),
            'league': league,
            'home': f['home_team'],
            'away': f['away_team'],
            'lambda_home': round(lam_h,3),
            'lambda_away': round(lam_a,3),
            'btts_prob': round(p,3),
            'home_recent_scored': league_info['teams'].get(f['home_team'],{}).get('scored', None),
            'away_recent_scored': league_info['teams'].get(f['away_team'],{}).get('scored', None)
        })
    df = pd.DataFrame(rows).sort_values('btts_prob', ascending=False).reset_index(drop=True)
    df.to_csv('btts_29nov2025_predictions.csv', index=False)
    print(df.head(40).to_string(index=False))
    print("\nSaved to btts_29nov2025_predictions.csv")
    return df

if __name__ == '__main__':
    run()
```
